```python
def generate_sample_data(num_samples=5):
    """Generate sample audio and MIDI files for training"""
    
    print("🎵 Generating sample data...")
    
    # Create directories
    os.makedirs('data/audio', exist_ok=True)
    os.makedirs('data/midi', exist_ok=True)
    
    def create_melody(base_notes, sr=22050, note_duration=0.5):
        """Create a melody from MIDI note numbers"""
        audio = np.array([])
        midi = pretty_midi.PrettyMIDI()
        piano = pretty_midi.Instrument(program=1)
        
        for i, note_num in enumerate(base_notes):
            # Generate sine wave
            freq = 440 * (2 ** ((note_num - 69) / 12))
            t = np.linspace(0, note_duration, int(sr * note_duration))
            note_audio = 0.3 * np.sin(2 * np.pi * freq * t)
            
            # Add envelope
            envelope = np.exp(-3 * t)
            note_audio *= envelope
            
            audio = np.concatenate([audio, note_audio])
            
            # Add to MIDI
            note = pretty_midi.Note(
                velocity=80,
                pitch=note_num,
                start=i * note_duration,
                end=(i + 1) * note_duration
            )
            piano.notes.append(note)
        
        midi.instruments.append(piano)
        return audio, midi, sr
    
    # Create different melodies
    melodies = [
        [60, 62, 64, 65, 67],  # C major scale
        [60, 64, 67, 72],      # C major chord
        [57, 60, 62, 65],      # A minor progression  
        [65, 64, 62, 60],      # Descending
        [60, 67, 64, 69, 62]   # Mixed pattern
    ]
    
    for i in range(num_samples):
        melody = melodies[i % len(melodies)]
        
        # Add variation
        if i > 0:
            shift = (i - 2) * 2  # Pitch shift
            melody = [note + shift for note in melody]
        
        audio, midi, sr = create_melody(melody)
        
        # Save files
        sf.write(f'data/audio/sample_{i:02d}.wav', audio, sr)
        midi.write(f'data/midi/sample_{i:02d}.mid')
    
    print(f"✅ Created {num_samples} sample files")
    return True

# Generate sample data
generate_sample_data(5)
```

```python
class AudioProcessor:
    """Handle audio loading and feature extraction"""
    
    def __init__(self, sr=22050, n_fft=2048, hop_length=512, n_mels=128):
        self.sr = sr
        self.n_fft = n_fft
        self.hop_length = hop_length
        self.n_mels = n_mels
        
    def load_audio(self, file_path):
        """Load audio file"""
        audio, sr = librosa.load(file_path, sr=self.sr)
        return audio
    
    def extract_melspectrogram(self, audio):
        """Extract mel-spectrogram features"""
        mel_spec = librosa.feature.melspectrogram(
            y=audio,
            sr=self.sr,
            n_fft=self.n_fft,
            hop_length=self.hop_length,
            n_mels=self.n_mels
        )
        log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)
        return log_mel_spec
    
    def visualize_audio(self, audio_path):
        """Visualize audio waveform and spectrogram"""
        audio = self.load_audio(audio_path)
        mel_spec = self.extract_melspectrogram(audio)
        
        fig, axes = plt.subplots(2, 1, figsize=(12, 8))
        
        # Waveform
        times = np.linspace(0, len(audio)/self.sr, len(audio))
        axes[0].plot(times, audio)
        axes[0].set_title('Audio Waveform')
        axes[0].set_xlabel('Time (s)')
        axes[0].set_ylabel('Amplitude')
        
        # Mel-spectrogram
        img = librosa.display.specshow(
            mel_spec, x_axis='time', y_axis='mel',
            sr=self.sr, hop_length=self.hop_length, ax=axes[1]
        )
        axes[1].set_title('Mel-spectrogram')
        plt.colorbar(img, ax=axes[1])
        
        plt.tight_layout()
        plt.show()
        
        return audio, mel_spec

# Test audio processing
audio_processor = AudioProcessor()
print("🎧 Audio processor created!")

# Visualize first sample
if os.path.exists('data/audio/sample_00.wav'):
    audio, mel_spec = audio_processor.visualize_audio('data/audio/sample_00.wav')
    print(f"Audio shape: {audio.shape}")
    print(f"Mel-spectrogram shape: {mel_spec.shape}")
```

```python
class MIDIProcessor:
    """Handle MIDI loading and conversion to piano roll"""
    
    def __init__(self, fs=100, min_note=21, max_note=108):
        self.fs = fs
        self.min_note = min_note
        self.max_note = max_note
        self.n_notes = max_note - min_note + 1
    
    def load_midi(self, file_path):
        """Load MIDI file"""
        midi = pretty_midi.PrettyMIDI(file_path)
        return midi
    
    def midi_to_piano_roll(self, midi):
        """Convert MIDI to piano roll"""
        piano_roll = midi.get_piano_roll(fs=self.fs)
        piano_roll = piano_roll[self.min_note:self.max_note+1, :]
        piano_roll = (piano_roll > 0).astype(np.float32)
        return piano_roll.T
    
    def align_with_audio(self, piano_roll, audio_frames):
        """Align piano roll with audio frames"""
        if piano_roll.shape[0] != audio_frames:
            time_ratio = audio_frames / piano_roll.shape[0]
            new_indices = np.arange(audio_frames) / time_ratio
            
            aligned_roll = np.zeros((audio_frames, piano_roll.shape[1]))
            for note_idx in range(piano_roll.shape[1]):
                aligned_roll[:, note_idx] = np.interp(
                    new_indices, 
                    np.arange(piano_roll.shape[0]), 
                    piano_roll[:, note_idx]
                )
            
            aligned_roll = (aligned_roll > 0.5).astype(np.float32)
            return aligned_roll
        
        return piano_roll
    
    def visualize_piano_roll(self, midi_path):
        """Visualize MIDI as piano roll"""
        midi = self.load_midi(midi_path)
        piano_roll = self.midi_to_piano_roll(midi)
        
        plt.figure(figsize=(12, 6))
        plt.imshow(piano_roll.T, aspect='auto', origin='lower', cmap='Blues')
        plt.colorbar(label='Note Active')
        plt.xlabel('Time Frames')
        plt.ylabel('MIDI Note Number')
        plt.title('Piano Roll Visualization')
        plt.show()
        
        return piano_roll
# Test MIDI processing
midi_processor = MIDIProcessor()
print("🎹 MIDI processor created!")

# Visualize first sample
if os.path.exists('data/midi/sample_00.mid'):
    piano_roll = midi_processor.visualize_piano_roll('data/midi/sample_00.mid')
    print(f"Piano roll shape: {piano_roll.shape}")
```

```python
class MusicDataset:
    """Handle loading and pairing of audio-MIDI data"""
    
    def __init__(self, audio_dir='data/audio', midi_dir='data/midi'):
        self.audio_dir = audio_dir
        self.midi_dir = midi_dir
        self.audio_processor = AudioProcessor()
        self.midi_processor = MIDIProcessor()
        self.file_pairs = self._get_file_pairs()
        
    def _get_file_pairs(self):
        """Get matching audio-MIDI file pairs"""
        if not os.path.exists(self.audio_dir) or not os.path.exists(self.midi_dir):
            return []
            
        audio_files = sorted([f for f in os.listdir(self.audio_dir) if f.endswith('.wav')])
        midi_files = sorted([f for f in os.listdir(self.midi_dir) if f.endswith('.mid')])
        
        pairs = []
        for audio_file in audio_files:
            base_name = os.path.splitext(audio_file)[0]
            midi_file = base_name + '.mid'
            
            if midi_file in midi_files:
                pairs.append((
                    os.path.join(self.audio_dir, audio_file),
                    os.path.join(self.midi_dir, midi_file)
                ))
        
        return pairs
    
    def load_sample(self, idx):
        """Load a single audio-MIDI pair"""
        audio_path, midi_path = self.file_pairs[idx]
        
        # Load audio
        audio = self.audio_processor.load_audio(audio_path)
        mel_spec = self.audio_processor.extract_melspectrogram(audio)
        
        # Load MIDI
        midi = self.midi_processor.load_midi(midi_path)
        piano_roll = self.midi_processor.midi_to_piano_roll(midi)
        
        # Align dimensions
        audio_frames = mel_spec.shape[1]
        piano_roll = self.midi_processor.align_with_audio(piano_roll, audio_frames)
        
        return mel_spec.T, piano_roll
    
    def get_all_data(self):
        """Load all data"""
        X, y = [], []
        
        for i in range(len(self.file_pairs)):
            try:
                mel_spec, piano_roll = self.load_sample(i)
                X.append(mel_spec)
                y.append(piano_roll)
            except Exception as e:
                print(f"Error loading sample {i}: {e}")
                continue
        
        return X, y
    
    def show_data_info(self):
        """Display dataset information"""
        print(f"📊 Dataset Information:")
        print(f"  Number of audio-MIDI pairs: {len(self.file_pairs)}")
        
        if len(self.file_pairs) > 0:
            mel_spec, piano_roll = self.load_sample(0)
            print(f"  Mel-spectrogram shape: {mel_spec.shape}")
            print(f"  Piano roll shape: {piano_roll.shape}")
            print(f"  Feature dimension: {mel_spec.shape[1]}")
            print(f"  Output dimension: {piano_roll.shape[1]}")

# Test dataset
dataset = MusicDataset()
dataset.show_data_info()
```

```python
def create_simple_model(input_dim=128, output_dim=88):
    """Create a simple transcription model"""
    
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(None, input_dim)),
        
        # Temporal convolution
        tf.keras.layers.Conv1D(64, 3, activation='relu', padding='same'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Conv1D(128, 3, activation='relu', padding='same'),
        tf.keras.layers.Dropout(0.2),
        
        # LSTM layers
        tf.keras.layers.LSTM(256, return_sequences=True, dropout=0.3),
        tf.keras.layers.LSTM(128, return_sequences=True, dropout=0.3),
        
        # Output
        tf.keras.layers.Dense(output_dim, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', 'precision', 'recall']
    )
    
    return model

# Create and display model
sample_mel, sample_piano = dataset.load_sample(0)
input_dim = sample_mel.shape[1]
output_dim = sample_piano.shape[1]

model = create_simple_model(input_dim, output_dim)
print("🤖 Model created successfully!")
model.summary()
```

```python
def pad_sequences(sequences, max_length=None):
    """Pad sequences to the same length"""
    if max_length is None:
        max_length = max(len(seq) for seq in sequences)
    
    padded = []
    for seq in sequences:
        if len(seq) < max_length:
            padding = np.zeros((max_length - len(seq), seq.shape[1]))
            padded_seq = np.vstack([seq, padding])
        else:
            padded_seq = seq[:max_length]
        padded.append(padded_seq)
    
    return np.array(padded)

def train_model_notebook(model, dataset, epochs=15):
    """Train the model in notebook environment"""
    
    print("🎯 Preparing training data...")
    
    # Load all data
    X, y = dataset.get_all_data()
    
    if len(X) == 0:
        print("❌ No data found!")
        return None
    
    # Pad sequences to same length
    X_padded = pad_sequences(X)
    y_padded = pad_sequences(y)
    
    print(f"📊 Data shapes - X: {X_padded.shape}, y: {y_padded.shape}")
    
    # Split train/validation
    n_samples = len(X_padded)
    train_size = max(1, int(0.8 * n_samples))
    
    X_train, X_val = X_padded[:train_size], X_padded[train_size:]
    y_train, y_val = y_padded[:train_size], y_padded[train_size:]
    
    # Handle case where we only have 1 sample
    if len(X_val) == 0:
        X_val, y_val = X_train[:1], y_train[:1]
    
    print(f"📊 Train: {len(X_train)} samples, Val: {len(X_val)} samples")
    
    # Callbacks
    callbacks = [
        tf.keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True, verbose=1),
        tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5, verbose=1)
    ]
    
    # Train
    print("🚀 Training started...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=min(2, len(X_train)),
        callbacks=callbacks,
        verbose=1
    )
    
    return model, history

# Train the model
model, history = train_model_notebook(model, dataset, epochs=10)

if model:
    # Save model in the correct format
    os.makedirs('models', exist_ok=True)
    try:
        # Try the new .keras format first
        model.save('models/music_transcriber.keras')
        print("💾 Model saved to models/music_transcriber.keras")
    except:
        # Fallback to .h5 format if .keras fails
        model.save('models/music_transcriber.h5')
        print("💾 Model saved to models/music_transcriber.h5")
```

```python
def plot_training_history(history):
    """Plot training metrics"""
    
    if history is None:
        print("No training history to plot")
        return
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Loss
    axes[0, 0].plot(history.history['loss'], label='Training Loss', marker='o')
    axes[0, 0].plot(history.history['val_loss'], label='Validation Loss', marker='s')
    axes[0, 0].set_title('Model Loss')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    
    # Accuracy
    axes[0, 1].plot(history.history['accuracy'], label='Training Accuracy', marker='o')
    axes[0, 1].plot(history.history['val_accuracy'], label='Validation Accuracy', marker='s')
    axes[0, 1].set_title('Model Accuracy')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Accuracy')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    
    # Precision
    axes[1, 0].plot(history.history['precision'], label='Training Precision', marker='o')
    axes[1, 0].plot(history.history['val_precision'], label='Validation Precision', marker='s')
    axes[1, 0].set_title('Model Precision')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Precision')
    axes[1, 0].legend()
    axes[1, 0].grid(True)
    
    # Recall
    axes[1, 1].plot(history.history['recall'], label='Training Recall', marker='o')
    axes[1, 1].plot(history.history['val_recall'], label='Validation Recall', marker='s')
    axes[1, 1].set_title('Model Recall')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Recall')
    axes[1, 1].legend()
    axes[1, 1].grid(True)
    
    plt.tight_layout()
    plt.show()
    
    # Print final metrics
    final_loss = history.history['val_loss'][-1]
    final_acc = history.history['val_accuracy'][-1]
    final_precision = history.history['val_precision'][-1]
    final_recall = history.history['val_recall'][-1]
    
    print(f"📊 Final Training Metrics:")
    print(f"  Validation Loss: {final_loss:.4f}")
    print(f"  Validation Accuracy: {final_acc:.4f}")
    print(f"  Validation Precision: {final_precision:.4f}")
    print(f"  Validation Recall: {final_recall:.4f}")

# Plot training results
plot_training_history(history)
```

```python
# Cell 10: Model Evaluation and Visualization
def evaluate_and_visualize(model, dataset):
    """Evaluate model and create visualizations"""
    
    print("📋 Evaluating model...")
    
    # Load test data
    X, y = dataset.get_all_data()
    
    if len(X) == 0:
        print("❌ No test data found!")
        return
    
    # Use last sample as test
    test_idx = -1
    mel_spec, true_piano_roll = X[test_idx], y[test_idx]
    
    # Predict
    input_data = mel_spec[np.newaxis, ...]
    prediction = model.predict(input_data, verbose=0)
    pred_piano_roll = prediction[0]
    
    # Binarize prediction
    pred_binary = (pred_piano_roll > 0.5).astype(np.float32)
    
    # Create visualization
    fig, axes = plt.subplots(4, 1, figsize=(15, 12))
    
    # Input mel-spectrogram
    axes[0].imshow(mel_spec.T, aspect='auto', origin='lower', cmap='viridis')
    axes[0].set_title('Input Mel-spectrogram')
    axes[0].set_ylabel('Mel Bins')
    
    # True piano roll
    axes[1].imshow(true_piano_roll.T, aspect='auto', origin='lower', cmap='Blues')
    axes[1].set_title('Ground Truth Piano Roll')
    axes[1].set_ylabel('MIDI Note')
    
    # Predicted probabilities
    axes[2].imshow(pred_piano_roll.T, aspect='auto', origin='lower', cmap='Reds', vmin=0, vmax=1)
    axes[2].set_title('Predicted Probabilities')
    axes[2].set_ylabel('MIDI Note')
    
    # Binary prediction
    axes[3].imshow(pred_binary.T, aspect='auto', origin='lower', cmap='Greens')
    axes[3].set_title('Binary Prediction (threshold=0.5)')
    axes[3].set_ylabel('MIDI Note')
    axes[3].set_xlabel('Time Frames')
    
    plt.tight_layout()
    plt.show()
    
    # Calculate metrics
    accuracy = np.mean(true_piano_roll == pred_binary)
    
    # Note-level analysis
    true_notes = np.sum(true_piano_roll, axis=0)
    pred_notes = np.sum(pred_binary, axis=0)
    
    notes_detected = np.sum((true_notes > 0) & (pred_notes > 0))
    total_true_notes = np.sum(true_notes > 0)
    
    print(f"📊 Evaluation Results:")
    print(f"  Frame-level accuracy: {accuracy:.3f}")
    print(f"  Notes correctly detected: {notes_detected}/{total_true_notes}")
    print(f"  Note detection rate: {notes_detected/max(total_true_notes,1):.3f}")
    
    return accuracy, pred_piano_roll, pred_binary

# Evaluate the model
if model:
    accuracy, pred_probs, pred_binary = evaluate_and_visualize(model, dataset)
```

```python
# Cell 11: Audio Transcription Function
def transcribe_audio_notebook(audio_path, model, output_path=None):
    """Transcribe an audio file to MIDI"""
    
    print(f"🎵 Transcribing {audio_path}...")
    
    # Process audio
    processor = AudioProcessor()
    audio = processor.load_audio(audio_path)
    mel_spec = processor.extract_melspectrogram(audio)
    
    # Predict
    input_data = mel_spec.T[np.newaxis, ...]
    prediction = model.predict(input_data, verbose=0)
    pred_piano_roll = prediction[0]
    
    # Convert to MIDI
    pred_binary = (pred_piano_roll > 0.5).astype(np.float32)
    
    midi = pretty_midi.PrettyMIDI()
    piano = pretty_midi.Instrument(program=1)
    
    fs = 100  # Frames per second
    min_note = 21  # A0
    
    notes_created = 0
    
    for note_idx in range(pred_binary.shape[1]):
        note_num = min_note + note_idx
        note_frames = pred_binary[:, note_idx]
        
        # Find note segments
        diff = np.diff(np.concatenate(([0], note_frames, [0])))
        onsets = np.where(diff == 1)[0]
        offsets = np.where(diff == -1)[0]
        
        for onset, offset in zip(onsets, offsets):
            start_time = onset / fs
            end_time = offset / fs
            
            if end_time - start_time > 0.05:  # Minimum duration
                note = pretty_midi.Note(
                    velocity=80,
                    pitch=note_num,
                    start=start_time,
                    end=end_time
                )
                piano.notes.append(note)
                notes_created += 1
    
    midi.instruments.append(piano)
    
    # Save MIDI
    if output_path is None:
        base_name = os.path.splitext(os.path.basename(audio_path))[0]
        output_path = f"transcribed_{base_name}.mid"
    
    midi.write(output_path)
    print(f"💾 Transcription saved to {output_path}")
    print(f"🎼 Created {notes_created} notes")
    
    # Visualize transcription
    fig, axes = plt.subplots(3, 1, figsize=(15, 10))
    
    # Original audio
    axes[0].plot(np.linspace(0, len(audio)/processor.sr, len(audio)), audio)
    axes[0].set_title('Original Audio Waveform')
    axes[0].set_ylabel('Amplitude')
    
    # Prediction probabilities
    axes[1].imshow(pred_piano_roll.T, aspect='auto', origin='lower', cmap='Reds', vmin=0, vmax=1)
    axes[1].set_title('Predicted Note Probabilities')
    axes[1].set_ylabel('MIDI Note')
    
    # Final transcription
    axes[2].imshow(pred_binary.T, aspect='auto', origin='lower', cmap='Blues')
    axes[2].set_title('Final Transcription (Binary)')
    axes[2].set_ylabel('MIDI Note')
    axes[2].set_xlabel('Time Frames')
    
    plt.tight_layout()
    plt.show()
    
    return output_path, notes_created

# Transcribe a sample audio file
if model and os.path.exists('data/audio/sample_00.wav'):
    transcribed_file, num_notes = transcribe_audio_notebook('data/audio/sample_00.wav', model)
    
    print(f"🎵 Transcription complete!")
    print(f"📁 You can play {transcribed_file} in any DAW or music software")

# Cell 12: Interactive Testing and Experimentation
print("🧪 Interactive Testing Section")
print("=" * 40)

# Test with different thresholds
def test_different_thresholds(model, dataset, thresholds=[0.3, 0.5, 0.7]):
    """Test model with different prediction thresholds"""
    
    # Get test data
    X, y = dataset.get_all_data()
    if len(X) == 0:
        return
    
    mel_spec, true_piano_roll = X[0], y[0]  # Use first sample
    
    # Predict
    input_data = mel_spec[np.newaxis, ...]
    prediction = model.predict(input_data, verbose=0)
    pred_piano_roll = prediction[0]
    
    # Test different thresholds
    fig, axes = plt.subplots(len(thresholds) + 1, 1, figsize=(15, 3*(len(thresholds)+1)))
    
    # Ground truth
    axes[0].imshow(true_piano_roll.T, aspect='auto', origin='lower', cmap='Blues')
    axes[0].set_title('Ground Truth')
    axes[0].set_ylabel('MIDI Note')
    
    # Different thresholds
    for i, threshold in enumerate(thresholds):
        pred_binary = (pred_piano_roll > threshold).astype(np.float32)
        accuracy = np.mean(true_piano_roll == pred_binary)
        
        axes[i+1].imshow(pred_binary.T, aspect='auto', origin='lower', cmap='Greens')
        axes[i+1].set_title(f'Threshold = {threshold} (Accuracy: {accuracy:.3f})')
        axes[i+1].set_ylabel('MIDI Note')
    
    axes[-1].set_xlabel('Time Frames')
    plt.tight_layout()
    plt.show()

# Test different thresholds
if model:
    test_different_thresholds(model, dataset)
```

```python
# Cell 13: Performance Analysis
def analyze_performance(model, dataset):
    """Detailed performance analysis"""
    
    print("📊 Detailed Performance Analysis")
    print("=" * 35)
    
    X, y = dataset.get_all_data()
    if len(X) == 0:
        return
    
    all_accuracies = []
    all_predictions = []
    all_true = []
    
    # Test on all samples
    for i in range(len(X)):
        mel_spec, true_piano_roll = X[i], y[i]
        
        # Predict
        input_data = mel_spec[np.newaxis, ...]
        prediction = model.predict(input_data, verbose=0)
        pred_piano_roll = prediction[0]
        pred_binary = (pred_piano_roll > 0.5).astype(np.float32)
        
        # Calculate accuracy
        accuracy = np.mean(true_piano_roll == pred_binary)
        all_accuracies.append(accuracy)
        
        # Store for overall metrics
        all_predictions.extend(pred_binary.flatten())
        all_true.extend(true_piano_roll.flatten())
    
    # Overall metrics
    overall_accuracy = np.mean(all_accuracies)
    precision, recall, f1, _ = precision_recall_fscore_support(
        all_true, all_predictions, average='binary'
    )
    
    print(f"📈 Performance Summary:")
    print(f"  Average accuracy across all samples: {overall_accuracy:.3f}")
    print(f"  Overall precision: {precision:.3f}")
    print(f"  Overall recall: {recall:.3f}")
    print(f"  F1-score: {f1:.3f}")
    
    # Per-sample breakdown
    print(f"\n📋 Per-sample accuracy:")
    for i, acc in enumerate(all_accuracies):
        print(f"  Sample {i}: {acc:.3f}")
    
    # Visualize performance distribution
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.hist(all_accuracies, bins=10, alpha=0.7, color='skyblue', edgecolor='black')
    plt.xlabel('Accuracy')
    plt.ylabel('Number of Samples')
    plt.title('Distribution of Sample Accuracies')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    metrics = ['Precision', 'Recall', 'F1-Score']
    values = [precision, recall, f1]
    bars = plt.bar(metrics, values, color=['lightcoral', 'lightgreen', 'lightblue'])
    plt.ylabel('Score')
    plt.title('Overall Performance Metrics')
    plt.ylim(0, 1)
    
    # Add value labels on bars
    for bar, value in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                f'{value:.3f}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()

# Run performance analysis
if model:
    analyze_performance(model, dataset)
```

```python
# Cell 15: Real-Time Transcription Simulation
def simulate_real_time_transcription(model, audio_path, window_size=2.0, overlap=0.5):
    """Simulate real-time transcription by processing audio in chunks"""
    
    print(f"⏱️  Simulating real-time transcription of {audio_path}")
    
    processor = AudioProcessor()
    audio = processor.load_audio(audio_path)
    sr = processor.sr
    
    # Calculate window parameters
    window_samples = int(window_size * sr)
    hop_samples = int(window_size * sr * (1 - overlap))
    
    all_predictions = []
    timestamps = []
    
    # Process audio in chunks
    start = 0
    while start + window_samples <= len(audio):
        # Extract audio chunk
        audio_chunk = audio[start:start + window_samples]
        
        # Extract features
        mel_spec = processor.extract_melspectrogram(audio_chunk)
        
        # Predict
        input_data = mel_spec.T[np.newaxis, ...]
        prediction = model.predict(input_data, verbose=0)
        pred_piano_roll = prediction[0]
        
        # Store results
        all_predictions.append(pred_piano_roll)
        timestamps.append(start / sr)
        
        start += hop_samples
    
    # Visualize real-time transcription
    if all_predictions:
        # Concatenate all predictions
        full_prediction = np.concatenate(all_predictions, axis=0)
        pred_binary = (full_prediction > 0.5).astype(np.float32)
        
        plt.figure(figsize=(15, 8))
        
        # Original audio
        plt.subplot(2, 1, 1)
        time_axis = np.linspace(0, len(audio)/sr, len(audio))
        plt.plot(time_axis, audio)
        plt.title('Original Audio')
        plt.ylabel('Amplitude')
        
        # Add window markers
        for i, ts in enumerate(timestamps):
            plt.axvline(x=ts, color='red', alpha=0.3, linestyle='--')
            if i < 5:  # Label first few windows
                plt.text(ts, max(audio)*0.8, f'W{i+1}', rotation=90)
        
        # Real-time transcription
        plt.subplot(2, 1, 2)
        plt.imshow(pred_binary.T, aspect='auto', origin='lower', cmap='Blues')
        plt.title('Real-time Transcription (Simulated)')
        plt.xlabel('Time Windows')
        plt.ylabel('MIDI Note')
        
        plt.tight_layout()
        plt.show()
        
        print(f"🎵 Processed {len(timestamps)} time windows")
        print(f"📊 Total transcription length: {full_prediction.shape[0]} frames")

# Test real-time simulation
if model and os.path.exists('data/audio/sample_00.wav'):
    simulate_real_time_transcription(model, 'data/audio/sample_00.wav')
```

```python
# Cell 17: Export and Deployment Functions
def export_model_for_deployment():
    """Export model in different formats for deployment"""
    
    if not model:
        print("❌ No model to export!")
        return
    
    print("📦 Exporting model for deployment...")
    
    # Create export directory
    os.makedirs('exports', exist_ok=True)
    
    try:
        # 1. Native Keras format (.keras)
        model.save('exports/music_transcriber.keras')
        print("✅ Keras format exported to exports/music_transcriber.keras")
        
        # 2. H5 format (legacy, but widely compatible)
        model.save('exports/music_transcriber.h5')
        print("✅ H5 format exported to exports/music_transcriber.h5")
        
        # 3. SavedModel format (for TensorFlow Serving)
        model.export('exports/music_transcriber_savedmodel')
        print("✅ SavedModel format exported to exports/music_transcriber_savedmodel/")
        
    except Exception as e:
        print(f"⚠️  Model export failed: {e}")
        # Fallback: just save as .keras
        try:
            model.save('exports/music_transcriber.keras')
            print("✅ Fallback: Keras format exported to exports/music_transcriber.keras")
        except Exception as e2:
            print(f"❌ All export methods failed: {e2}")
            return
    
    # 4. TensorFlow Lite (for mobile/edge deployment)
    try:
        converter = tf.lite.TFLiteConverter.from_keras_model(model)
        # Add optimizations for better mobile performance
        converter.optimizations = [tf.lite.Optimize.DEFAULT]
        tflite_model = converter.convert()
        
        with open('exports/music_transcriber.tflite', 'wb') as f:
            f.write(tflite_model)
        print("✅ TensorFlow Lite model exported to exports/music_transcriber.tflite")
        
        # Get model size
        tflite_size = len(tflite_model) / 1024 / 1024
        print(f"   📊 TFLite model size: {tflite_size:.2f} MB")
        
    except Exception as e:
        print(f"⚠️  TensorFlow Lite export failed: {e}")
    
    # 5. Model summary and metadata
    try:
        with open('exports/model_info.txt', 'w') as f:
            model.summary(print_fn=lambda x: f.write(x + '\n'))
            f.write(f"\nModel Parameters:\n")
            f.write(f"Input Shape: (None, {input_dim})\n")
            f.write(f"Output Shape: (None, {output_dim})\n")
            f.write(f"Total Parameters: {model.count_params()}\n")
            f.write(f"Model Memory (approx): {model.count_params() * 4 / 1024 / 1024:.2f} MB\n")
        
        print("✅ Model info exported to exports/model_info.txt")
    except Exception as e:
        print(f"⚠️  Model info export failed: {e}")
    
    # 6. Create deployment script template
    deployment_script = '''#!/usr/bin/env python3
"""
Music Transcription Deployment Script
Usage: python deployment_script.py your_audio.wav
"""

import tensorflow as tf
import numpy as np
import librosa
import pretty_midi
import sys
import os

class MusicTranscriber:
    def __init__(self, model_path='music_transcriber.keras'):
        """Initialize transcriber with trained model"""
        try:
            self.model = tf.keras.models.load_model(model_path)
            print(f"✅ Model loaded from {model_path}")
        except Exception as e:
            print(f"❌ Failed to load model: {e}")
            print("Available model files:")
            for ext in ['.keras', '.h5']:
                if os.path.exists(f'music_transcriber{ext}'):
                    print(f"  - music_transcriber{ext}")
            return None
            
        self.sr = 22050
        self.n_mels = 128
    
    def transcribe(self, audio_path, output_path=None):
        """Transcribe audio file to MIDI"""
        
        # Load and process audio
        audio, sr = librosa.load(audio_path, sr=self.sr)
        mel_spec = librosa.feature.melspectrogram(
            y=audio, sr=sr, n_mels=self.n_mels
        )
        log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)
        
        # Predict
        input_data = log_mel_spec.T[np.newaxis, ...]
        prediction = self.model.predict(input_data, verbose=0)
        pred_piano_roll = prediction[0]
        
        # Convert to MIDI
        pred_binary = (pred_piano_roll > 0.5).astype(np.float32)
        
        midi = pretty_midi.PrettyMIDI()
        piano = pretty_midi.Instrument(program=1)
        
        fs = 100  # Frames per second
        min_note = 21  # A0
        
        notes_created = 0
        for note_idx in range(pred_binary.shape[1]):
            note_num = min_note + note_idx
            note_frames = pred_binary[:, note_idx]
            
            # Find note segments
            diff = np.diff(np.concatenate(([0], note_frames, [0])))
            onsets = np.where(diff == 1)[0]
            offsets = np.where(diff == -1)[0]
            
            for onset, offset in zip(onsets, offsets):
                start_time = onset / fs
                end_time = offset / fs
                
                if end_time - start_time > 0.05:  # Minimum duration
                    note = pretty_midi.Note(
                        velocity=80,
                        pitch=note_num,
                        start=start_time,
                        end=end_time
                    )
                    piano.notes.append(note)
                    notes_created += 1
        
        midi.instruments.append(piano)
        
        # Save MIDI
        if output_path is None:
            base_name = os.path.splitext(os.path.basename(audio_path))[0]
            output_path = f"transcribed_{base_name}.mid"
        
        midi.write(output_path)
        print(f"🎵 Transcription saved to {output_path} ({notes_created} notes)")
        
        return output_path

# Command line interface
if __name__ == "__main__":
    if len(sys.argv) < 2:
        print("Usage: python deployment_script.py <audio_file> [output_file]")
        sys.exit(1)
    
    audio_file = sys.argv[1]
    output_file = sys.argv[2] if len(sys.argv) > 2 else None
    
    # Initialize transcriber
    transcriber = MusicTranscriber()
    if transcriber.model is None:
        sys.exit(1)
    
    # Transcribe
    try:
        result = transcriber.transcribe(audio_file, output_file)
        print(f"✅ Success! Check {result}")
    except Exception as e:
        print(f"❌ Transcription failed: {e}")
'''
    
    try:
        with open('exports/deployment_script.py', 'w') as f:
            f.write(deployment_script)
        
        # Make it executable on Unix systems
        import stat
        st = os.stat('exports/deployment_script.py')
        os.chmod('exports/deployment_script.py', st.st_mode | stat.S_IEXEC)
        
        print("✅ Deployment script created at exports/deployment_script.py")
    except Exception as e:
        print(f"⚠️  Deployment script creation failed: {e}")
    
    # 7. Create requirements file for deployment
    requirements = """tensorflow>=2.13.0
librosa>=0.10.1
pretty_midi>=0.2.10
soundfile>=0.12.1
numpy>=1.24.3
"""
    
    try:
        with open('exports/requirements.txt', 'w') as f:
            f.write(requirements)
        print("✅ Requirements file created at exports/requirements.txt")
    except Exception as e:
        print(f"⚠️  Requirements file creation failed: {e}")
    
    print(f"\n📦 Export Summary:")
    export_files = []
    for filename in ['music_transcriber.keras', 'music_transcriber.h5', 
                     'music_transcriber.tflite', 'model_info.txt', 
                     'deployment_script.py', 'requirements.txt']:
        filepath = f'exports/{filename}'
        if os.path.exists(filepath):
            size = os.path.getsize(filepath) / 1024 / 1024
            export_files.append(f"  📁 {filename} ({size:.2f} MB)")
    
    if export_files:
        print("Exported files:")
        for file_info in export_files:
            print(file_info)
    else:
        print("❌ No files were exported successfully")
    
    if os.path.exists('exports/music_transcriber_savedmodel'):
        print("  📁 music_transcriber_savedmodel/ (directory)")
    
    print(f"\n🚀 Deployment Ready!")
    print(f"  Use: python exports/deployment_script.py your_audio.wav")

# Export model
if model:
    export_model_for_deployment()

# Cell 18: Performance Optimization Tips
print("⚡ Performance Optimization Tips")
print("=" * 35)

def analyze_model_complexity():
    """Analyze model complexity and suggest optimizations"""
    
    if not model:
        return
    
    total_params = model.count_params()
    trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    
    print(f"🔍 Model Complexity Analysis:")
    print(f"  Total parameters: {total_params:,}")
    print(f"  Trainable parameters: {trainable_params:,}")
    print(f"  Model size (approx): {total_params * 4 / 1024 / 1024:.2f} MB")
    
    # Timing test
    dummy_input = tf.random.normal((1, 250, input_dim))
    
    # Warm up
    _ = model.predict(dummy_input, verbose=0)
    
    # Time prediction
    import time
    start_time = time.time()
    for _ in range(10):
        _ = model.predict(dummy_input, verbose=0)
    avg_time = (time.time() - start_time) / 10
    
    print(f"  Average inference time: {avg_time:.3f} seconds")
    print(f"  Throughput: {1/avg_time:.1f} predictions/second")
    
    # Optimization suggestions
    print(f"\n💡 Optimization Suggestions:")
    if total_params > 1000000:
        print("  - Consider reducing model size for faster inference")
    if avg_time > 0.1:
        print("  - Model might be too slow for real-time applications")
        print("  - Consider quantization or pruning")
    
    print("  - Use TensorFlow Lite for mobile deployment")
    print("  - Enable GPU acceleration if available")
    print("  - Use batch processing for multiple files")

analyze_model_complexity()
```
